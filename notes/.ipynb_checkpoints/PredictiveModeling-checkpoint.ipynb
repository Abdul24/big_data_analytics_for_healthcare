# Predictive Modeling 
a process of modeling historical data for predicting future events

## Predictive Modeling Pipeline
1. Prediction Target
2. Cohort Construction
3. Feature Construction
4. Create Predictive Model
5. Evaluate Model

## Prediction Target
We will focus on predicting heart failure, which lies in the sweet spot of both interesting, and possible to solve problems.
![img](./img/pred_targ.png)

### Motivations For Early Detection of Heart Failure
1. Heart failure is a complex disease
2. Reduce costs and hospitilizations
3. Slow disease progression
4. Improve existing clinical guidelines for preventing heart failur

### Cohort Construction (study design)
* There exists a subset of the population for which our study will be valid.
* We should keep in mind that we probably won't have data for this entire subset, but our data should come from this subset.

![img](./img/study_pop.png)
#### Prospective vs. Retrospective
* Prospective: Identify patients, then begin data collection
* Retrospective: Identify patients, then retrieve all historical data about these patients

![img](./img/prospective_v_retrospective.png)

### Cohort Study
* Select a group of patients who are exposed to the risk.
    * Target: Heart Failure Readmission
        * Cohort: all HF patients discharged from hospital
        * Key: define the right inclusion/exclusion criteria

### Case-Control Study
* Cases: Patients with positive outcomes
* Controls: Patients with negative outcome (but otherwise similar to _cases_)
![img](./img/case_control.png)

### Feature Construction
![img](./img/feature_timeline.png)
* We use the observation window to construct features.
* The length of the `observation window` and `prediction window` can heavily influence model performance.
    * Large observation window and small prediction window are easiest to model.
    * Small observation window, and large prediction window is the most useful (powerful) model, but most difficult to construct.
    
| Prediction Window | Observation Window |
|-------------------|--------------------|
| ![img](./img/prediction_window.png) | ![img](./img/observation_window.png) |

### Feature Selection
Identify predictive features from a large set of possible features.

### Predictive Model
* Regression
* Classification

### Performance Evaluation
* Training error is not useful
* Test on unseen samples
* Approach: Cross Validation

#### Cross Validation
* Leave 1 Out
* K-Fold
* Random
